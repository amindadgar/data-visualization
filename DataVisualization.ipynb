```python
def get_box_plot_data_from_dataframe(df, column, groupby):
  """
  Get an array of data for a column group by another column

  Parameters:
  -------------
  df : dataframe
    the dataset available
  column : string
    the column that wanted to be returned
    note that this column must be in `df`
  gropuby : string
    the column that the dataset is grouped by
    note that this column must be in `df`

  Returns:
  ---------
  boxPlot_data : array
    an array of one column that is grouped by another column in `df` dataset
  """
  boxPlot_data = []
  ## get the region available in the dataset
  unique_groupedBy_column = df1[groupby].unique()

  ## get each ladder scores
  for col in unique_groupedBy_column:
    boxPlot_data.append(df1[df1[groupby] == col][column])

  return boxPlot_data
```

```python
U_regional_data = get_box_plot_data_from_dataframe(df1, 'Ladder score', 'Regional indicator')

plt.figure(figsize=(15, 10))
## box plot for each region ladder score
plt.boxplot(U_regional_data)
plt.xticks( np.linspace(1,10, 10) ,df1['Regional indicator'].unique(), rotation=30)
plt.show()
```

It seems that `Western Europe` and `North America and ANZ` has the highest wellness and happiness in the world because its box is placed in the highest range.

The lowest values belong to `South Asia` and `Sub-Saharan Africa` while its box is places in the lowest place. But there is an important thing to note that, Middle East and North Africa can also placed in this lowest category since its mean is placed in a low place (the orange line is the mean).

While `Western Europe` and `North America and ANZ` are the best regions according to `Ladder Score`, Let's see how freedom to make life choices happen to be. 

```python
choicesFreedom_data = get_box_plot_data_from_dataframe(df1, 'Freedom to make life choices', 'Regional indicator' )


plt.figure(figsize=(15, 10))
## box plot for each region ladder score
plt.boxplot(choicesFreedom_data)
plt.xticks( np.linspace(1,10, 10) ,df1['Regional indicator'].unique(), rotation=30)
plt.show()
```

Well, It seems again `Western Europe` and `North America and ANZ` also have the highest value for freedom of life choices. The surprising thing here that the region `Southeast Asia` also have the highest value for freedom of life choices.
